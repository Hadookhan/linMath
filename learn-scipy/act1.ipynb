```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp # scipy ODE solver function

# Define right side of the ODE as function
def dydt(t, y):
    return -2 * y

t_span = (0, 5) # tells solver where to start and stop
y0 = [1] # initial state as a sequence

# calls solver with dydt (passes in time and y values)
# t_span=(0,5) integrates from 0 to 5
# y0=[1] is the initial condition
# t_eval=np.linespace(0,5,100) requests solution at 100 evenly spaced times between 0 and 5
sol = solve_ivp(dydt, t_span, y0, t_eval=np.linspace(0, 5, 100))

# sol.t refers to the times of sol object
# sol.y refers to the solution values with shape (n_vars, n_time). Here (1, 100). Solution at y[0]
plt.plot(sol.t, sol.y[0], label='Numerical Solution')
# e-2t is the known solution for the ODE
plt.plot(sol.t, np.exp(-2*sol.t), 'r--', label='Exact Solution')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.title('ODE Solution')
plt.show()
```
