```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits import mplot3d

# creates an 3d projection of an axes
ax = plt.axes(projection="3d")
ax.scatter(3,5,7)
plt.show()
```

```python
ax = plt.axes(projection='3d')

# randomizes data sample for x, y, z
x_data = np.random.randint(0,100, (500,))
y_data = np.random.randint(0,100, (500,))
z_data = np.random.randint(0,100, (500,))

# uses scatter graph to plot data
ax.scatter(x_data, y_data, z_data, marker="v")

plt.show()
```

```python
ax = plt.axes(projection='3d')

x_data = np.arange(0, 50, 0.1)
y_data = np.arange(0, 50, 0.1)
z_data = np.sin(x_data)*np.cos(y_data)

# plot gives linear graph
ax.plot(x_data, y_data, z_data)
ax.set_title("3d graph")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```

```python
ax = plt.axes(projection='3d')

x_data = np.arange(-5, 5, 0.1)
y_data = np.arange(-5, 5, 0.1)

# meshgrid creates coordinate matrices from coordinate vectors
X, Y = np.meshgrid(x_data, y_data)
Z = np.sin(X) * np.cos(Y)

# cmap allows for a gradient on the surface object
ax.plot_surface(X, Y ,Z, cmap="plasma")
plt.show()
```

```python
ax = plt.axes(projection='3d')

x_data = np.arange(-5, 5, 0.1)
y_data = np.arange(-5, 5, 0.1)

# meshgrid creates coordinate matrices from coordinate vectors
X, Y = np.meshgrid(x_data, y_data)
Z = np.sin(X) * np.cos(Y)

# cmap allows for a gradient on the surface object
ax.plot_surface(X, Y ,Z, cmap="plasma")
# changes original viewpoint of graph
ax.view_init(azim=0, elev=90)
plt.show()
```
